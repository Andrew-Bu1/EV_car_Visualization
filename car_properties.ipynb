```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
car_properties = pd.read_csv('data/car_properties/EV_cars_properties.csv')
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(car_properties['Range'], car_properties['Price.DE.'], c=np.arctan2(car_properties['Range'], car_properties['Price.DE.']), cmap='rainbow', s=50, alpha=0.8)

plt.title('Electric Vehicles: Range vs Price')
plt.xlabel('Range (km)')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```

```python

# Create a pivot table
pivot_table = car_properties.pivot(index="Car_name", columns="Fast_charge", values="Range")
# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", linewidths=.5)
plt.title('EV Battery Charge Time vs Range')
plt.xlabel('Battery Charge Time (hours)')
plt.ylabel('Car Model')
plt.show()
```
