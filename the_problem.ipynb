```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
transport_percentage = pd.read_csv("data/the_problem/transport_emission.csv")
```

```python
plt.figure(figsize=(10, 7))
sns.set_palette("pastel")
plt.pie(transport_percentage["Percentage"], labels=transport_percentage["Entity"], autopct='%1.1f%%', startangle=140)
plt.title('Percentage Distribution of Different Vehicles')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
